Description of the data in each column:

*   ***work_year***	- the year in which the respective salary was paid
*   ***experience_level*** -	the level of experience of the respective person, with 4 possible values:
     -    EN Entry-level / Junior 
     -    MI Mid-level / Intermediate 
     -    SE Senior-level / Expert 
     -    EX Executive-level / Director
*   ***employment_type*** -	type of employment: 
    - PT - Part-time
    - FT - Full-time
    - CT - Contract
    - FL - Freelance

*   ***job_title*** -	the position the person worked on
*   ***salary*** -	total gross salary
*   ***salary_currency*** -	the currency in which the salary was paid (according to the ISO 4217 currency code standard)
*   ***salaryinusd*** -	salary in dollars (according to the exchange rate of the day)
*   ***employee_residence*** - the employee's country of residence (according to the ISO 3166 country code standard)
*   ***remote_ratio***	the percentage of remote work, with the possible values:
    - 0 - No remote work (less than 20%) 
    - 50 - Partially remote 
    - 100 - Fully remote (more than 80%)
*   ***company_location*** -	company location (country according to ISO 3166 country code)
*   ***company_size*** -	calculated according to the average number of people who worked in the company in the respective type of year:
      -  S less than 50 employees (small) 
      -  M between 50 and 250 employees (medium) 
      -  L more than 250 employees (large)

## Define an **extractColumn()** function to extract all data that refers to a column from the dataset. The function will have 3 input parameters:

*  all_data_lst = the list containing all the data (without header), obtained in the previous point
*  col_name = the name of the column to be extracted from the data
*  header = the list containing the names of the columns, obtained in the previous point

```python
def extractColumn(my_data_lst, col_name, header):
  col_idx = header.index(col_name)
  result = []
  for line in my_data_lst:
    result.append(line[col_idx])
  return result

```

## Write a function that converts all elements of a list to ints

For example, for the list of String elements:

*   ["79833", "260000", "109024", "20000", "150000"]

the result will be a list of elements of type int (note the lack of quotes):
 
*   [79833, 260000, 109024, 20000, 150000]


```python
def transform_int(my_list):
  salaries = []
  for x in my_list:
      x = int(x)
      salaries.append(x)
  return salaries

```

## Show for each experience level (Entry, middle, senior, executive) what is the average salary in a bar chart

```python
def extractSalaryForExpLevel(all_data, col_name, header, queried_level):
  col_idx = header.index(col_name)
  exp_level_idx = header.index("experience_level")

  result = []
  for line in all_data:
    exp_level = line[exp_level_idx]
    if exp_level == queried_level:
      result.append(line[col_idx])
  return result
```

```python
EN_level = extractSalaryForExpLevel(all_data, 'salary_in_usd', header, 'EN')
MI_level = extractSalaryForExpLevel(all_data, 'salary_in_usd', header, 'MI')
SE_level = extractSalaryForExpLevel(all_data, 'salary_in_usd', header, 'SE')
EX_level = extractSalaryForExpLevel(all_data, 'salary_in_usd', header, 'EX')


d1 = {"EN": sum(transform_int(EN_level)),
      "MI": sum(transform_int(EN_level)),
      "SE": sum(transform_int(EN_level)),
      "EX": sum(transform_int(EN_level))}

EN_ave = round(sum(transform_int(EN_level)) / 88 , 2 )
MI_ave = round(sum(transform_int(EN_level)) / 213 , 2)
SE_ave = round(sum(transform_int(EN_level)) / 280 , 2)
EX_ave = round(sum(transform_int(EN_level)) / 26 , 2)

print(d1)
print("Media la nivelul Entry", EN_ave)
print("Media la nivelul Middle", MI_ave)
print("Media la nivelul Senior", SE_ave)
print("Media la nivelul Executive", EX_ave) 
```

## Define the **extractColumnForYear(all_data_lst, col_name, header, year)** function, which allows extracting data for a column from the data set, **only if the 'work_year' column for that row also coincides with year**

```python
def extractColumnForYear(all_data, col_name, header, year):
  col_idx = header.index(col_name)
  year_col_idx = header.index("work_year")

  result = []
  for line in all_data:
    year_data = line[year_col_idx]
    if year_data == year:
      result.append(line[col_idx])
  return result
```

```python
work_year2020 = extractColumnForYear(all_data, 'salary_in_usd', header, '2020')
work_year2021 = extractColumnForYear(all_data, 'salary_in_usd', header, '2021')
work_year2022 = extractColumnForYear(all_data, 'salary_in_usd', header, '2022')
```
